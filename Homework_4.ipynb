### 1.1. Скачать данные по ссылке 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import piplite
await piplite.install('seaborn')
import seaborn as sns
```

```python
df = pd.read_csv('kc-house-data.csv', sep=',')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(8, 6))
plt.hist(df['price'], range=(0, 3000000), bins= 20)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Количество домов');
```

Большая часть домов продаётся в ценовом диапазоне от 100 до 900 тысяч долларов.

Самая распространённая цена за дом 400-500 тысяч долларов.

### 1.3 Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(8, 6))
plt.hist(df['sqft_living'], range=(0, 8000), bins= 20)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('sqft')
plt.ylabel('Количество домов');
```

Большинство объектов недвижимости имеют квадратуру от примерно 1000 до 3000 квадратных футов.

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(8, 6))
plt.hist(df['yr_built'], bins= 20)
plt.title('Распределение по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов');
plt.locator_params (axis='x', nbins= 20)
```

Большинство объектов недвижимости были построены в период с 1946 по 2010-е годы.

```python
plt.hist(df['floors'], bins=15)
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов');
```

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
names = data.index
values = data.values
plt.figure(figsize=(5, 3))
plt.bar(names, values)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов');
```

Состояние домов довольно хорошее. 

Большинство домов имеют оценку от 3 до 5. 

Больше всего домов с оценкой 3.

## Задание 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
corr_matrix = df.corr(numeric_only = True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

На стоимость недвижимости влияют следующие характеристики:
- жилая площадь
- оценка дома
- площадь без подвала
- кол-во спален
- кол-во ванных комнат
- оценка вида
- кол-во этажей
- наличие набережной
- площадь подвала
- жилая площадь 15 ближайших соседей

```python
df_sorted = df.sort_values('grade')
sns.boxplot(x=df['price'], y=df_sorted['grade'].astype('str'), showfliers=False)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))
plt.xlabel('Цена')
plt.ylabel('оценка дома')
plt.title('Взаимосвязь стоимости дома и его оценки');
```

Чем выше оценка дома, тем выше его стоимость и разброс цены.

```python
df_sorted = df.sort_values('floors')
sns.boxplot(x=df['price'], y=df_sorted['floors'].astype('str'), showfliers=False)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))
plt.xlabel('Цена')
plt.ylabel('Количество этажей')
plt.title('Взаимосвязь стоимости дома и оценки вида');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5.

```python
df_sorted = df.sort_values('grade')
sns.boxplot(x=df['price'], y=df_sorted['view'].astype('str'), showfliers=False)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))
plt.xlabel('Цена')
plt.ylabel('оценка вида')
plt.title('Взаимосвязь стоимости дома и оценки вида');
```

Чем выше оценка вида, тем выше стоимость дома.

Дома с оценкой вида 1-2 имеют примерно одинаковые значения стоимости.

```python
df_sorted = df.sort_values('bedrooms')
sns.boxplot(x=df['price'], y=df_sorted['bedrooms'].astype('str'), showfliers=False)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))
plt.xlabel('Цена')
plt.ylabel('Количество спален')
plt.title('Взаимосвязь стоимости дома и количества спален');
```

Стоимость домов возрастает с увеличением количества спален.

Дома с 2-3 и с 5-7 спальнями имеют примерно одну стоимость.

Дома с 8 спальнями имеют самый большой разброс цен.

```python
plt.figure(figsize=(10, 8))
df_sorted = df.sort_values('bathrooms')
sns.boxplot(x=df['price'], y=df_sorted['bathrooms'].astype('str'), showfliers=False)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))
plt.xlabel('Цена')
plt.ylabel('Количество ванных комнат')
plt.title('Взаимосвязь стоимости дома и количества ванных комнат');
```

Стоимость домов возрастает с увеличением количества ванных комнат.

Максимальную стоимость имеют дома с 8 ваннами комнатами.
